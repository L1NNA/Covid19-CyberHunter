```python
import requests
```

```python
key = 'your_key'
```

```python
t1 = 'https://maps.googleapis.com/maps/api/place/findplacefromtext/json?input={}&inputtype=textquery&fields=place_id,formatted_address,name,geometry&key={}'
t2 = 'https://maps.googleapis.com/maps/api/place/details/json?place_id={}&key={}'
```

```python
data = {}
```

```python
from tqdm import tqdm
queries = list(set(locs))

for q in tqdm(queries):
    if q in data:
        continue
    search = requests.get(t1.format(q, key)).json()
    if len(search['candidates']) >0 :
        top = search['candidates'][0]
        _id = top['place_id']
        _name = top['name']
        search = requests.get(t2.format(_id, key)).json()
        if 'website' in search['result']:
            _web = search['result']['website']
        else:
            _web = ''
        _prov = search['result']['address_components'][-3]['long_name']
        data[q] = (_prov, _web, _name, q, _id)
    else:
        print('not found ' + q)
    
```
